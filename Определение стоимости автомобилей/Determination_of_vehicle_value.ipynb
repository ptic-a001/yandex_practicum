```python
def explore_dataframe_text(df):
    
    # выводим первые 5 строк датафрейма
    print("*" * 50)
    print("Первые 5 строк датафрейма:")
    print("*" * 50)
    display(df.head())
    print()
    
    # выводим информацию о типах данных и количестве пропущенных значений
    print("*" * 50)
    print("Информация о данных:")
    print("*" * 50)
    print()
    print(df.info())
    print()
    
    # выводим информацию о типах данных и количестве пропущенных значений
    print("*" * 50)
    print("Информация % Nan:")
    print("*" * 50)
    print()
    display(pd.DataFrame(round(df.isna().mean()*100)).rename({0:'% Nan'},axis=1).style.background_gradient('coolwarm'))
    print()
    
    
    # выводим описание некоторых статистических характеристик 
    print("*" * 50)
    print("Статистические характеристик : ")
    print("*" * 50)
    print()
    display(pd.concat([df.isnull().sum(), # есть ли nan
           df.dtypes,         # типы
           
           df.nunique(), # кол-во уникальных значений
           df.describe().T], axis=1).rename({0:'nan', 1:'type', 2:'n_unique'},axis=1))
    print()
    
   
    
    # выводим все дублирующиееся строки и их кол-во
    
    print("*" * 50)
    print("Вывод числа дублированных данных")
    print("*" * 50)
    
    print(df.duplicated().sum())
    
    print("*" * 50)
    print("Вывод дублированных данных")
    print("*" * 50)
    
    display(df[df.duplicated(keep=False)])
    
    print()
    

    
    
    #выводим корреляцию между признаками 
    print("*" * 50)
    print("Корреляция между признаками ")
    print("*" * 50)
    
    plt.figure(figsize=(12,8))
    display(sns.heatmap(df.corr(),cmap="RdBu"))
    display(plt.show())
    print()
```

```python
def drow_stat_lines(df):
    plt.vlines(min(df), ymin = 0, ymax=14500, color='r',linewidth=3)

    plt.vlines(df.median(), ymin = 0, ymax=14500, color='tab:purple',linewidth=3)
    plt.vlines(df.mean(), ymin = 0, ymax=14500, color='g',linewidth=3)
    plt.vlines(df.quantile(0.25), ymin = 0, ymax=14500, color='k',linewidth=3)
    plt.vlines(df.quantile(0.75), ymin = 0, ymax=14500, color='k',linewidth=3)
```

```python
plt.figure(figsize=(15,10))
sns.histplot(x=df['date_created'].sort_values(), data=df, palette="Set1",bins = 1000)
plt.grid()
drow_stat_lines(df['date_created'])
plt.title('Распределение дат создания анкет')
plt.ylabel('Количество созданных анкет')
plt.xlabel('Дата создания анкеты');
```

```python
q1 = df['date_created'].quantile(0.25)
q3 = df['date_created'].quantile(0.75)
iqr = q3 - q1
x = df['date_created'].apply(lambda x: df['date_created'].median() if x < q1 - 1.5*iqr or x > q3 + 1.5*iqr else x)
plt.figure(figsize=(15,10))
sns.histplot(x=x.sort_values(), data=df, palette="Set1",bins = 100, alpha=0.3)
drow_stat_lines(x)
plt.title('Распределение дат создания анкет')
plt.ylabel('Количество созданных анкет')
plt.xlabel('Дата создания анкеты')
plt.grid();
```

## Обучение моделей

Трансформеры + моделька линейной регрессии с L2-регуляризацией 

```python
scaler = StandardScaler()
oh_encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
o_encoder = OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = 251)

model = Ridge()
```

Разбивка на признаки

```python
cols = df.dtypes.reset_index()

nums = df[cols['index']].nunique().reset_index().drop('index', axis=1)

num_columns = cols[cols[0].map(lambda x: (x in ['int32', 'int8', 'int64', 'float64']))]['index'] # здесь числовые
num_columns = num_columns[num_columns!='price'].to_list()

cat_columns = cols[(cols[0] == 'object') & (nums[0] <= 10)]['index'].to_list() # здесь категориальные, с числом уникальны < 10

cat_ordinal_columns = cols[(cols[0]=='object') & (nums[0]>=10)]['index'].to_list()#здесь категориальные, с числом уникальных >10
```

```python
seed = 0xEDA
X = df.drop(['price'], axis=1)
y = df['price']

# делю на трейн и тест
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

print(X_train.shape, X_test.shape)
```

**Валидация**

```python
nfold = 3 
seed = 0xEDA 
gkf = list(GroupKFold(n_splits=nfold).split(df_train, groups=df_train['date_created']))
```

```python
lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'n_estimators': 150,
    'learning_rate': 0.09,    
}
```

```python
drop_cols = ['date_created', 'price']

features = df_train.drop(drop_cols, 1, errors='ignore').columns
target = 'price'
```

```python

print(f'Number of features: {len(features)}')
oof = np.zeros(len(df_train))
oof_lr = np.zeros(len(df_train))
train_preds = np.zeros(len(df_train))
train_preds_lr = np.zeros(len(df_train))
df_train = df_train.drop('date_created', axis=1)

full_time = {'lgbm_time' : 0,'lr_time': 0}
models = []
models_lr = []

for fold_, (train_idx, val_idx) in enumerate(gkf):
    best_score = 10**6
    for i in range(2):
        print(f'Training with fold {fold_} started.')
        lgb_params['learning_rate'] -= i/50
        lgb_model = lgb.LGBMRegressor(**lgb_params)
        

        df_train_, df_val_ = df_train.iloc[train_idx], df_train.iloc[val_idx]
        df_train_ = pd.DataFrame(pipeline['columntransformer'].fit_transform(df_train_), columns = df_train_.columns)
        df_val_ = pd.DataFrame(pipeline['columntransformer'].fit_transform(df_val_), columns = df_val_.columns)
        start_time = time.time()
        lgb_model.fit(df_train_[features], 
                      df_train_[target].values, 
                      eval_set=[(df_val_[features], df_val_[target].values)],
                      verbose=10)
        print(f'Time taken for lgbm: {time.time() - start_time:.2f} seconds')
        full_time['lgbm_time'] += time.time() - start_time




        
        if mean_squared_error(df_val_["price"], lgb_model.predict(df_val_[features]), squared=False) < best_score:
            
            oof[val_idx] = lgb_model.predict(df_val_[features])
            train_preds[train_idx] += lgb_model.predict(df_train_[features]) / (nfold-1)
            models.append(lgb_model)
            best_score = mean_squared_error(df_val_["price"], oof[val_idx], squared=False)
            print(f'Training with fold {fold_} completed.')

    lgb_params['learning_rate'] += 1/50
    
target = 'price'
features = df_train.drop('price', axis=1).columns
for fold_, (train_idx, val_idx) in enumerate(gkf):
    lr_model = pipeline2['ridge']
    df_train_, df_val_ = df_train.iloc[train_idx], df_train.iloc[val_idx]
    print(df_train_.shape)
    print(df_val_.shape)
    
        # Обучаем объект ColumnTransformer на тренировочном датасете
    pipeline2['columntransformer'].fit(df_train_[features])
    
    # Применяем метод transform() вместо fit_transform() к объекту ColumnTransformer
    df_2 = pd.DataFrame(pipeline2['columntransformer'].transform(df_train_[features]).toarray())
    df2_val = pd.DataFrame(pipeline2['columntransformer'].transform(df_val_[features]).toarray())
    
    df_2 = df_2.reset_index(drop=True)
    df2_val = df2_val.reset_index(drop=True)
    
    ser1 = df_train_[target]
    ser2 = df_val_[target]
    ser1 = ser1.reset_index(drop=True)
    ser2 = ser2.reset_index(drop=True)
    
    print(df_2.shape)
    print(df2_val.shape)
    

    

    
    #df_2.columns = df_2.columns.astype(str)
    #df2_val.columns = df2_val.columns.astype(str)
    
    start_time = time.time()
    lr_model.fit(df_2, 
                  ser1)
    print(f'Time taken for lr: {time.time() - start_time:.2f} seconds')
    full_time["lr_time"] += time.time() - start_time

    oof_lr[val_idx] = lr_model.predict(df2_val)
    models_lr.append(lr_model)
    print(f'Training with fold {fold_} completed.')
```
